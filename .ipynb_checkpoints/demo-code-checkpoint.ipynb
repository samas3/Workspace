### 反馈系统 - 声音

```python
import os
from pydub import AudioSegment
import simpleaudio as sa

# 加载 MP3 文件
audio = AudioSegment.from_mp3("test.mp3")
# 导出为 WAV 文件
audio.export("test.wav", format="wav")

# ==== 播放 WAV 文件 ====

# 设置 ALSA 环境变量指定音频设备
os.environ['ALSA_CARD'] = '3'
os.environ['ALSA_DEVICE'] = '0'

# 播放音频文件
wave_obj = sa.WaveObject.from_wave_file("demo1.wav")
play_obj = wave_obj.play()
play_obj.wait_done()
```

### 反馈系统 - 快动作

```python
from adafruit_servokit import ServoKit

kit=ServoKit(channels=16)
servo=3

# 舵机引脚配置
servoB = kit.servo[0] # Reference at 90
servoL = kit.servo[1] # Reference at 180
servoR = kit.servo[2] # Reference at 0
    
servoB.angle = 0
servoL.angle = 90
servoR.angle = 90


servoB.angle = 90
servoR.angle = 180
servoL.angle = 180
```

### 反馈系统 - 慢动作

```python
while(True):
    for i in range(0, 12):
        if i <= 3:
            servoR.angle = 90 + i*10 #at 120
            servoL.angle = 90 - i*10 #at 60
            servoB.angle = 90 - i*10
        if (i > 3 and i <=9):
            servoR.angle = 150 - i*10 #at 60
            servoL.angle = i*10 + 30 #at 120
            servoB.angle = i*10 + 30
        if i>9:
            servoR.angle = i*10 - 30 #at 90
            servoL.angle = 210 - i*10 #at 90
            servoB.angle = 210 - i*10
        time.sleep(0.4)
```

### 反馈系统 - 显示GIF表情

```python
# 选择你想显示的表情文件：”dizzy.gif”, “happy.gif”, "sad.gif", "sleep.gif", "blink.gif"
gif = Image.open(dir_path + '/dizzy.gif')
frame_count = gif.n_frames # 提取 GIF 图像帧数

# 把每一帧保存图片都存下来
for i in range(frame_count):
    gif.seek(i)
    rgb_frame = gif.convert("RGB")
    rgb_frame.save(f"frame_{i:04d}.png", "PNG")

# 显示每一帧图片
for i in range(frame_count):
    image = Image.open(home_dir+"/"+f"frame_{i:04d}.png")
    disp.ShowImage(image)
```

# 感知系统

### 感知系统 - 听觉（麦克风）

```python
import sounddevice as sd
from scipy.io.wavfile import write
import numpy as np

fs = 44100 # 采样频率
seconds = 5 # 录音时长，单位：秒
output_filename = "output.wav" # 保存录制的音频为 "自定义文件名.wav"

print("* 开始录音")
myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=1) # 录制音频
sd.wait() # 等待录音完成
print("* 录音结束")

# 将浮点格式转换为整数格式
myrecording = np.int16(myrecording / np.max(np.abs(myrecording)) * 32767)

# 保存音频文件
write(output_filename, fs, myrecording)
print(f"音频已保存为 {output_filename}")
```

### 感知系统 - 触摸（传感器）

```python
import lgpio

# 告诉系统："我要开始使用GPIO了"
h = lgpio.gpiochip_open(0)

# 引脚配置
touch_pin = 17
vibration_pin = 22

# 相当于告诉硬件："这两个引脚用于接收信号"
lgpio.gpio_claim_input(h, touch_pin) # 将17号引脚配置为输入模式
lgpio.gpio_claim_input(h, vibration_pin) 
```

```python
touch = lgpio.gpio_read(h, touch_pin) # 读取17号引脚的当前电平值（0/1）
vibration = lgpio.gpio_read(h, vibration_pin)
if touch:
    print("Touch detected")
if vibration:
    print("Vibration detected")
```
